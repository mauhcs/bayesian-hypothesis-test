# Binomial distribution

Given an experiment that has only two possible outcomes $\{ 0, 1\}$, if the probablility of the outcome 1 is equal to $\theta \in [0,1]$, given $N$ trials of the experiment the probability of occurring $z$ events is given by the $binomial$ distribution with the following form:

$$bin(z | N, \theta) = {N \choose z} \theta^z (1-\theta)^{N-z}$$


Since it is easier to talk about a concrete example instead of trials and events of abstract events, let's stick to tosses of a coin as a concrete example. In this case, the trials will be each separate toss, and the outcomes \{0, 1}\ will be `tails` and `heads` of a coin toss. In this manner, the equation for the binomial distribution can be read as the propability of $z$ heads in $N$ tosses.

Let's do a Python example using `scipy`'s default functions

```python
# Import scipy
from scipy.stats import binom

# Say we perform 26 coin tosses and there were 8 heads, i.e, N=26 and z=8
N = 29
z = 10
theta = 0.5
# The probaility is given by binom.pmf(8, N, p):
print(f"bin(z|N,theta)={binom.pmf(z, N, theta)}")
```

**Note** that when calling the scupy default function `bin` to calculate the binomial distribution you noticed that we used the `pmf` method and it stands for *Probability Mass Function*. This name is just to make it clear that the probability function we are using is a discrete function (instead of a continous function).




$H_0:$ The coin is fair

$H_A:$ The coins ia biased (unfair)

```python
from dsbay.distribution import DistributionMass
x = [_z for _z in range(N+1)]
y = [binom.pmf(_z, N, theta) for _z in range(N+1)]

d = DistributionMass(x,y)
d.plot()
```
